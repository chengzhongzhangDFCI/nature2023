```python
def segment_and_track_cell (images, 
                            cell_centroid = [176,103],
                            cell_area = 7000,
                            time = 0, 
                            z_selected = 1,
                            nuclei_filter = 600,
                            crop = True):

    
    
    # load 3d images for the timepoint selected
    image_dna = []
    for _lyr in range(images.sizes["z"]):
        lyr_dna = np.array(images.get_frame_2D (v=time, c=0, z=_lyr))[850:1190,850:1250]
        image_dna.append(lyr_dna)
        
    image_dna = np.array(image_dna)

    image_lac = []
    for _lyr in range(images.sizes["z"]):
        lyr_lac = np.array(images.get_frame_2D (v=time, c=1, z=_lyr))[850:1190,850:1250]
        image_lac.append(lyr_lac)
        
    image_lac = np.array(image_lac)  

    image_ms2 = []
    for _lyr in range(images.sizes["z"]):
        lyr_ms2 = np.array(images.get_frame_2D (v=time, c=2, z=_lyr))[850:1190,850:1250]
        image_ms2.append(lyr_ms2)
        
    image_ms2 = np.array(image_ms2)  
    
    # segment DAPI and track the cell
    
    model= models.Cellpose(gpu=False, model_type='nuclei')

    dna_masks, flows, styles, diams = model.eval(image_dna[z_selected], diameter=None, channels = [0,0],
                                         flow_threshold=0.4, do_3D=False)
    
    print(np.unique(dna_masks))
    dna_masks =  remove_small_objects(dna_masks, 100 ,connectivity=1)
    dna_masks = erosion(dna_masks, disk(20))
    dna_masks = dilation(dna_masks, disk(20))
    print(np.unique(dna_masks))
    
    found_dist = 600
    
    
    num_of_nuclei = len(np.unique(dna_masks)) - 1
    for i in range(num_of_nuclei):
        cand_nucleus = dna_masks == i+1
        cand_nucleus[cand_nucleus>0]=1
        cand_nucleus = np.array(cand_nucleus)
        
        
        
        region = skimage.measure.regionprops (skimage.measure.label(cand_nucleus))[0]
        
        _dist = np.linalg.norm(np.array([region.centroid[1], region.centroid[0]])
                                                    - np.array(cell_centroid))
        
        if _dist < found_dist:
            if abs(region.area - cell_area)/cell_area <=0.2:
                found_dist = _dist
                kept_nucleus = cand_nucleus
    
    
    region = skimage.measure.regionprops (skimage.measure.label(kept_nucleus))[0]    
    found_cell_centroid = [region.centroid[1], region.centroid[0]]
    found_cell_area = region.area
    
    return [kept_nucleus, found_cell_centroid, found_cell_area, found_dist]
    
    

    
```
